### Q2.b-What is the total number of matches won by Sunrises Hyderabad?

```python
df.loc[df["winner"]=="Sunrisers Hyderabad", 'winner'].value_counts()
```

### Q2.c-What is the ratio of toss decision - field .vs. bat?

```python
df.loc[df["toss_decision"]=="field", "toss_decision"].count()/df.loc[df["toss_decision"]=="bat", "toss_decision"].count()
```
